```python
import pandas as pd
```

```python
df = pd.read_excel('/content/datos_prod_aguacate_hass_2001_2022 (1).xlsx')
df.head()
```

```python
df.dtypes
```

```python
#Definimos los conjuntos X e y
X = df[['Volumen_Ton']]
y = df['Valor_MXN'] #Millones de Pesos
y = y.round(4)
print(X.head())
print('------------------------------')
print(y.head())
```

```python
print(X.isnull().sum())
print(y.isnull().sum())
```

```python
#Dividimos los datos en conjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2, random_state=1
)#20% de los datos para prueba
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Puedes ajustar el max_depth
tree_regressor = DecisionTreeRegressor(random_state=42, max_depth=4)
tree_regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Predicción en el conjunto de prueba
y_pred = tree_regressor.predict(X_test)

# Cálculo del error cuadrático medio y r^2
mse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Square Error:", rmse)
print("R² Score:", r2)
```

```python
nuevas_ofertas = pd.DataFrame({
    'Volumen_Ton' : [3, 120, 1600, 18000, 220000, 2400000]
})

predicciones = tree_regressor.predict(nuevas_ofertas)
print(predicciones)
```
